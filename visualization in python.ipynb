```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python

# Create a date range
dates = pd.date_range(start="2020-01-01", periods=100, freq='B')  # 'B' frequency is for business days

# Generate a DataFrame to simulate US stock market data with five related variables
np.random.seed(0)  # For reproducibility
data = {
    'Open': np.random.uniform(100, 200, size=len(dates)),  # Simulated opening price
    'High': np.random.uniform(200, 300, size=len(dates)),  # Simulated daily high price
    'Low': np.random.uniform(50, 100, size=len(dates)),    # Simulated daily low price
    'Close': np.random.uniform(150, 250, size=len(dates)),  # Simulated closing price
    'Volume': np.random.randint(100000, 1000000, size=len(dates))  # Simulated trading volume
}

# Ensure 'High' is greater than 'Open', 'Close', and 'Low'; 'Low' is less than 'Open' and 'Close'
for i in range(len(dates)):
    data['High'][i] = max(data['Open'][i], data['High'][i], data['Close'][i])
    data['Low'][i] = min(data['Open'][i], data['Low'][i], data['Close'][i])

df_stock = pd.DataFrame(data, index=dates)

# Print the first few rows of the DataFrame
df_stock.head()
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Define the dimensions of the plot grid
n_vars = iris.shape[1] - 1  # Number of variables (excluding the species column)
fig, axes = plt.subplots(n_vars, n_vars, figsize=(12, 12))

# Plot each pair of variables
for i in range(n_vars):
    for j in range(n_vars):
        if i == j:
            # Diagonal: Plot a histogram
            axes[i, j].hist(iris.iloc[:, i], bins=15, color='gray', alpha=0.7)
        else:
            # Off-diagonal: Plot a scatter plot
            axes[i, j].scatter(iris.iloc[:, j], iris.iloc[:, i], alpha=0.7)

        # Set axis labels
        if j == 0:
            axes[i, j].set_ylabel(iris.columns[i])
        if i == n_vars - 1:
            axes[i, j].set_xlabel(iris.columns[j])

        # Remove axis labels for non-edge plots
        if i != n_vars - 1:
            axes[i, j].set_xlabel('')
        if j != 0:
            axes[i, j].set_ylabel('')

# Adjust layout
plt.tight_layout()
plt.show()
```

Joint Plot:

```python
plt.style.use('ggplot')  # Apply the 'ggplot' style to the plots for a certain aesthetic (similar to R's ggplot2)
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)  # Compute the sine of each x value

plt.plot(x, y)  # Plot y vs. x as lines
plt.annotate('Local Max', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))  # Annotate a point with text and an arrow
plt.show()  # Display the plot
```

# Seaborn: Facet Grid

```python
tips = sns.load_dataset("tips")  # Load the 'tips' dataset from Seaborn's dataset collection
g = sns.FacetGrid(tips, col="time", row="smoker")  # Create a FacetGrid. Each subplot will correspond to a combination of 'time' and 'smoker'
g.map(plt.hist, "total_bill")  # Map a histogram to each subplot based on the 'total_bill' column
plt.show()  # Display the plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(plt.hist, "total_bill")

# Iterate over each subplot (ax) in the FacetGrid and set the same y-axis limits
max_count = tips['total_bill'].value_counts().max()
g.set(ylim=(0, max_count))

plt.show()
```

# Plotly Subplots

Subplots allow you to display multiple plots in a single figure, which is useful for comparing different but related datasets.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Creating a subplot layout with 1 row and 2 columns
fig = make_subplots(rows=1, cols=2)  # Initialize a figure with subplots

# Adding a scatter plot to the first subplot
fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[4, 5, 6]),  # Scatter plot data
    row=1, col=1  # Position in the subplot grid
)

# Adding a bar plot to the second subplot
fig.add_trace(
    go.Bar(x=[1, 2, 3], y=[6, 5, 4]),  # Bar plot data
    row=1, col=2  # Position in the subplot grid
)

fig.show()  # Display the figure
```

# Seaborn Advanced Features

PairGrid (Customizable Pairplot):

Seaborn's PairGrid allows for more control compared to pairplot. It lets you specify which plots to use on the diagonal, upper half, and lower half of the grid.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

iris = sns.load_dataset('iris')

# Initialize the PairGrid
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)  # Upper triangle with scatter plots
g.map_lower(sns.kdeplot)  # Lower triangle with KDE plots
g.map_diag(sns.histplot)  # Diagonal with histograms
g.add_legend()  # Add a legend

plt.show()
```

# Clustermap (Advanced Heatmap):

Seaborn's clustermap is used for hierarchically-clustered heatmaps, which are great for visualizing complex matrices.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(color_codes=True)  # Set the theme for Seaborn plots

# Load and pivot the sample data
flights = sns.load_dataset("flights")
flights_pivoted = flights.pivot(index="month", columns="year", values="passengers")  # Corrected pivot call

# Draw a clustermap
sns.clustermap(flights_pivoted, cmap="mako", standard_scale=1)  # Create a clustermap
plt.show()  # Display the plot
```

# Matplotlib Advanced Features

3D Plotting:
Matplotlib can create 3D plots, which are useful for visualizing three-dimensional data.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional scatter
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)
ax.scatter(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

# Customizing Plots with Matplotlib:

Matplotlib's real power lies in its ability to be highly customizable. This includes modifying plot elements, adding annotations, and working with axes.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='--', color='green', marker='o', markersize=5)
plt.title('Customized Sin Wave')
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.grid(True)
plt.fill_between(x, y, color='green', alpha=0.1)

plt.show()
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd

# Load and pivot the sample data
flights = sns.load_dataset("flights")
flights_pivoted = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap with Plotly
fig = px.imshow(flights_pivoted, 
                labels=dict(x="Year", y="Month", color="Passengers"),
                x=flights_pivoted.columns,
                y=flights_pivoted.index,
                title="Flight Passengers Heatmap")
fig.show()
```
